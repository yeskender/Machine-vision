```python
#FIRST MATRIX
# (6) poorly removes noise such as "salt and pepper"
# (3) increases the sharpness of the image
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib.image as img
img = img.imread('./tank.png')
k = np.array([[1/16,1/8,1/16],
              [1/8,1/4,1/8],
              [1/16,1/8,1/16]])
flt = ndimage.convolve(img, k, mode='nearest', cval=0.0)
fmin = flt.min()
fmax = flt.max()
flt = (flt-fmin)/(fmax-fmin)

fig = plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(flt,cmap='gray')
plt.subplot(132)
plt.imshow(img,cmap='gray')
plt.subplot(133)
plt.imshow((img-flt),cmap='gray')
```

```python
#SECOND MATRIX
# (1) corresponds to a low-pass filter in the frequency domain;
# (3) increases the sharpness of the image

from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib.image as img
img = img.imread('./tank.png')
k = np.array([[0,1/5,0],[1/5,1/5,1/5],[0,1/5,0]])
flt = ndimage.convolve(img, k, mode='nearest', cval=0.0)
fmin = flt.min()
fmax = flt.max()
print(fmin)
print(fmax)
flt = (flt-fmin)/(fmax-fmin)
fig = plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(flt,cmap='gray')
plt.subplot(132)
plt.imshow(img,cmap='gray')
plt.subplot(133)
plt.imshow((img-flt),cmap='gray')
```

```python
#MEDIAN FILTER
# (2) less than the other specified noise-canceling filters, blurs edges;
# (6) poorly removes noise such as "salt and pepper"; 

from scipy import ndimage as nimg
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
imlen = plt.imread('./tank.png')

gflt = nimg.median_filter(imlen, 10)
fig = plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(gflt, cmap = 'gray')
plt.subplot(122)
plt.imshow(imlen, cmap = 'gray')

plt.show()
```

```python
#FOURTH MATRIX
# (5) is not a noise canceling filter
# (1) corresponds to a low-pass filter in the frequency domain; 
# (4) works slower than the other specified noise-canceling filters;
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib.image as img
img = img.imread('./tank.png')
k = np.array([[-1,-2,-1],[-2,12,-2],[-1,-2,-1]])
flt = ndimage.convolve(img, k, mode='nearest', cval=0.0)
fmin = flt.min()
fmax = flt.max()
flt = (flt-fmin)/(fmax-fmin)
fig = plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(flt,cmap='gray')
plt.subplot(132)
plt.imshow(img,cmap='gray')
plt.subplot(133)
plt.imshow((img-flt),cmap='gray')
```
