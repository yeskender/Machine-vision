```python
#MEDIAN FILTER
import cv2
from scipy import ndimage as nimg
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
imlen = plt.imread('./tank.png')

gflt = nimg.median_filter(imlen, 10)
fig = plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(gflt, cmap = 'gray')
plt.subplot(122)
plt.imshow(imlen, cmap = 'gray')

plt.show()
```

```python
#Median filter
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
from scipy.ndimage.filters import median_filter
from matplotlib.pylab import imread, imshow, subplot, title, show, get_cmap
from skimage import feature
im = pltimg.imread('./tank.png')
fig = plt.figure(figsize=(10,10))
median = median_filter(im, 5)
fig = plt.figure(figsize=(15,10))
plt.subplot(131)
plt.axis('off')
plt.title('Median filter', fontsize=20)
plt.imshow(median,cmap='gray')
plt.subplot(132)
plt.axis('off')
plt.title('Noisy image', fontsize=20)
plt.imshow(im,cmap='gray')
plt.subplot(133)
plt.axis('off')
plt.title('Difference', fontsize=20)
plt.imshow((im-median),cmap='gray')
plt.show();
```
